```python
terms_in_doc = json.load(open("terms_in_doc.json","r"))
```

```python
# Import generic wrappers
from transformers import AutoModel, AutoTokenizer 


# Define the model repo
model_name = "nlpaueb/legal-bert-base-uncased" 


# Download pytorch model
model = AutoModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)


```

```python
def get_embedding(term):
    # Transform input tokens 
    inputs = tokenizer(term, return_tensors="pt")

    # Model apply
    outputs = model(**inputs)
    
    return outputs
```

```python
cols = ['tid']
for i in range(768):
    cols.append(str(i))
bert_features = pd.DataFrame(columns=cols)
```

```python
bert_features
```

```python
count = 0
for i in terms_in_doc.keys():
    terms = terms_in_doc[i]
    n = len(terms)
    res = []
    
    if n != 0:
        for j in range(n):
            res.append(get_embedding(terms[j])[0][0][0])
    g = [0 for i in range(768)]
    g = np.array(g)
    for k in range(n):
        g = g + (res[k].detach().numpy())
    
    if n!=0:
        g = g/n
    
    row = [int(i)]
    for l in range(len(g)):
        row = np.append(row,g[l])
    len(row)
    bert_features.loc[count] = row
    count = count + 1
```
