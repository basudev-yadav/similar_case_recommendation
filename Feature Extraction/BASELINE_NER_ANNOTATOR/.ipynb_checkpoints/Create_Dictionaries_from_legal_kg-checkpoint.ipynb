```python
def max_token(tid,dictionary,label):
    dict_count = {}
    
    X = pd.read_csv("tokens_per_tid/"+str(tid)+".csv")
    
    for i in range(X.shape[0]):
        if X.loc[i,'label'] == label:
            if X.loc[i,'token'] in dict_count.keys():
                dict_count[X.loc[i,'token']] = dict_count[X.loc[i,'token']] + 1
            else:
                dict_count[X.loc[i,'token']] = 1
                
    
    #Keymax = max(zip(dict_count.values(), dict_count.keys()))[1]
    #return dictionary[Keymax]

    sorted_tuples = sorted(dict_count.items(), key=lambda item: item[1])
    sorted_dict = {k: v for k, v in sorted_tuples}
    
 
    if len(dict_count) > 0:
        if label == 'location' or label == 'organization':
            if len(dict_count) > 1:
                return dictionary[list(sorted_dict.keys())[0]] , dictionary[list(sorted_dict.keys())[1]]
            else:
                return dictionary[list(sorted_dict.keys())[0]] , 0
        else:
            return dictionary[list(sorted_dict.keys())[0]]
    else:
        if label == 'location' or label == 'organization':
            return 0,0
        else:
            return 0
```
