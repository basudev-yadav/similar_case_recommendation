```python
start = 16

for i in range(start,start+5):
    
        file = files[i]
        f = open(file,"r",encoding="utf-8")

        text = f.read()
        
        
        #nltk package
        
        for sent in nltk.sent_tokenize(text):
            for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent))):
                 if hasattr(chunk, 'label'):
                    nltk_labels.append(chunk.label()) 
            
        ########################################################################
        
        
        #spacy package
        
        nlp = spacy.load("en_core_web_sm")
        doc = nlp(text)
        
        for ent in doc.ents:
            spacy_labels.append(ent.label_)
            

        ########################################################################
        
        
        #stanza package
        
        nlp = stanza.Pipeline(lang='en', processors='tokenize,ner')
        doc = nlp(text)
        
        for ent in doc.ents:
            stanza_labels.append(ent.type)
            
        
        
```

```python
all_labels = []
```

```python

for i in range(len(files)):
    
        file = files[i]
        f = open(file,"r",encoding="utf-8")

        text = f.read()
        
        labels_spacy = []
        
        #spacy package
        
        nlp = spacy.load("en_core_web_sm")
        doc = nlp(text)
        
        for ent in doc.ents:
            labels_spacy = np.append(labels_spacy , ent.label_)
            all_labels = np.append(all_labels , ent.label_)
            
        labels_spacy = np.unique(labels_spacy)
        all_labels = np.unique(all_labels)
        
        regex = re.compile(r'[0-9]+')
        
        tid = regex.search(file).group()
        
        data = [tid , " , ".join(labels_spacy)]
        
        ent_df.loc[i] = data
        
```
