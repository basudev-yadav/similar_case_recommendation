```python
import pandas as pd
import numpy as np
```

```python
bert = pd.read_csv("BERT.csv")
ner_law = pd.read_csv("NER_LAW.csv")
law = pd.read_csv("LAWPOINTS.csv")
domain = pd.read_csv("DOMAIN_EXPERT.csv")
```

```python
len(bert),len(ner_law),len(law),len(domain)
```

```python
bert_nodes = bert['tid']
ner_law_nodes = ner_law['tid']
law_nodes = law['tid']
domain_nodes = domain['tid']
```

```python
set(bert_nodes) == set(ner_law_nodes) == set(law_nodes) == set(domain_nodes)
```

```python
#sort all the feature sets by IDs
```

```python
bert = bert.sort_values(by=['tid'])
ner_law = ner_law.sort_values(by=['tid'])
law = law.sort_values(by=['tid'])
domain = domain.sort_values(by=['tid'])
```

```python
bert
```

```python
bert = bert.reset_index()


law = law.reset_index()


ner_law = ner_law.reset_index()


domain = domain.reset_index()
```

```python
domain.drop(['index'],axis=1,inplace=True)
law.drop(['index'],axis=1,inplace=True)
ner_law.drop(['index'],axis=1,inplace=True)
bert.drop(['index'],axis=1,inplace=True)
```

```python
domain.to_csv("DOMAIN_EXPERT.csv",index=False)
law.to_csv("LAWPOINTS.csv",index=False)
ner_law.to_csv("NER_LAW.csv",index=False)
bert.to_csv("BERT.csv",index=False)
```
