```python
def jaccard_binary(x,y):
    """A function for finding the similarity between two binary vectors"""
    intersection = np.logical_and(x, y)
    union = np.logical_or(x, y)
    similarity = intersection.sum() / float(union.sum())
    return similarity
```

```python
features = pd.read_csv("OE_LABELS_NEW_FEATURES.csv")
```

```python
feature_test = features[features['tid'] == int(test_ikanoon_id)]
feature_test = feature_test.values.flatten().tolist()[1:]
```

```python
s_score = 0
most_similar = 0
scores = {}
for i in range(features.shape[0]):
    if features.iloc[i,0] != int(test_ikanoon_id):
        feature_i = features.loc[i]
        feature_i = feature_i.values.flatten().tolist()[1:]
        s = jaccard_binary(feature_test , feature_i)
        scores[features.iloc[i,0]] = s
        if s > s_score:
            s_score = s
            most_similar = features.iloc[i,0]
```
