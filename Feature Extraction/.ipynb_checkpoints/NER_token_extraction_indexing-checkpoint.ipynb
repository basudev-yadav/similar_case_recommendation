```python
tokens = []
for i in range(len(files)):
    #read file
    if files[i].endswith(".csv"):
        file = pd.read_csv(files[i],encoding="utf-8")
        for i in range(file.shape[0]):
            if file.iloc[i,1] != "CARDINAL":
                tokens.append(file.iloc[i,2])
```

```python
tokens = np.unique(tokens)
```

```python
for i in range(len(tokens)):
    tokens[i] = remove_punctuation(tokens[i])
```

```python
tokens = np.unique(tokens)
```

```python
len(np.unique(tokens))
```

```python
for i in range(len(tokens)):
    tokens[i] = " ".join(tokens[i].split())
    
tokens = np.unique(tokens)
```

```python
import json
```

```python
token_dict = {}
for i in range(len(tokens)):
    token_dict[i] = tokens[i]
```

```python
len(token_dict)
```

```python
fw = open("token_dict.json","w")
json.dump(token_dict,fw,indent=4)
```

```python
import string
```

```python
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree.lower()
```
