```python
import pandas as pd
import re
import os
import numpy as np
```

```python
data = pd.read_csv("casemine.csv")
```

```python
pd.set_option('display.max_rows',1000)
```

```python
names = data["Judgement Name"]
```

```python
names = np.unique(names)
```

```python
names
```

```python
for judgement in names:
    
    similar_list = []
    judgement_id = ""
    
    for i in range(data.shape[0]):
        if data.iloc[i,0] == judgement:
            if not pd.isna(data.iloc[i,3]):
                similar_list.append(data.iloc[i,3])
            judgement_link = data.iloc[i,1]
            if pd.isna(judgement_link):
                judgement_id = judgement
            else:
                regex = re.compile(r'[0-9]+')
                res = regex.search(judgement_link)
                judgement_id = res.group()
        
    os.mkdir(os.getcwd() + str('\\') + judgement_id)
    
    for j in similar_list:
        regex = re.compile(r'[0-9]+')
        res = regex.search(j)
        f = open(os.getcwd() + str('\\') + judgement_id + str("\\") + res.group(),"w")
        f.close()
    
    
```
