```python
import networkx as nx
```

```python
f = open("edge_samples.txt","r")
```

```python
G = nx.Graph()
```

```python
for line in f.readlines():
    u = int(line.split(',')[0])
    v = int(line.split(',')[1])
    G.add_edge(u,v)
```

```python
nx.draw(G, with_labels=True)
```

```python
nx.write_gexf(G,"sample_graph.gexf")
```
