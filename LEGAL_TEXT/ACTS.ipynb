```python
import pandas as pd
import numpy as np
import re
```

```python
import os
```

```python
for file in os.listdir():
    if file.endswith(".txt"):
        print(file)
```

```python
acts_with_dot = []
acts_without_dot = []


for file in os.listdir():
    if file.endswith(".txt"):
        
        if file:
            f = open(file,"r",encoding="utf-8")
            text = f.read()

            regex1 = re.compile(r'([A-Z]\.' '+)+ Act')
            regex2 = re.compile(r'([A-Z]' '+ Act)')

            res1 = regex1.search(text)
            res2 = regex2.search(text)

            if res1:
                acts_with_dot.append(res1.group())

            if res2:
                acts_without_dot.append(res2.group())
            
            
```

```python
acts_with_dot = np.unique(acts_with_dot)
acts_without_dot = np.unique(acts_without_dot)
```

```python
acts_with_dot
```

```python
acts_without_dot
```

```python
temp = []
```

```python
for act in acts_with_dot:
    
    x = re.sub(r'\.','',act)
    temp.append(x)
    
temp = np.unique(temp)
```

```python
temp
```

```python
acts = [i for i in temp]
for i in acts_without_dot:
    acts.append(i)
```

```python
acts = np.unique(acts)
acts
```

```python
acts_dict = {}
for i in acts:
    acts_dict[i] = ''
```

```python
acts_dict
```

```python
import json
```

```python
json_file = json.dumps(acts_dict)
```

```python
out_file = open("acts_dictionary.json","w")
json.dump(acts_dict,out_file)
out_file.close()
```

```python
dict_json = open("acts_dictionary.json")
```

```python
acts_d = json.load(dict_json,encoding="utf-8")
```

```python
acts_d
```

```python
acts_csv = pd.DataFrame(columns=["Abbreviated Form","Full Form"])
```

```python
counter = 0

for i in acts_d.keys():
    val = acts_d[i]
    vals = val.split(" ")
    t = []
    for j in vals:
        t.append(j.capitalize())
    
    val = " ".join(t)
    row = [i,val]
    acts_csv.loc[counter] = row
    counter = counter + 1
```

```python
acts_csv.to_csv("Abbreviated Form Dictionary.csv",index=False)
```
