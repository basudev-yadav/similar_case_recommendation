```python
#Prepare feature matrix for NER + lawpoints
```

```python
import pandas as pd
import numpy as np
import json
import re
import os
```

```python
mapping = json.load(open("MAPPING.json","r"))
```

```python
ft = pd.read_csv("FINAL_OE_LABELS_WITH_NER.csv")
```

```python
ft
```

```python
for i in range(ft.shape[0]):
    ft.iloc[i,0] = mapping[str(ft.iloc[i,0])]
```

```python
ft
```

```python
ft = ft.sort_values(by=['tid'])
```

```python
ft
```

```python
ft = ft.reset_index()
```

```python
ft
```

```python
ft.drop(['index'],axis=1,inplace=True)
```

```python
ft
```

```python
X = np.zeros((1038,28))
```

```python
for i in range(ft.shape[0]):
    row = ft.loc[i].values.flatten().tolist()
    row = row[1:]
    for j in range(len(row)):
        X[i][j] = row[j]
```

```python
X
```

```python
X
```

```python
import scipy.sparse
```

```python
X_sp = scipy.sparse.csc_matrix(X)
```

```python
scipy.sparse.save_npz('X_NER_LAW.npz',X_sp)
```
