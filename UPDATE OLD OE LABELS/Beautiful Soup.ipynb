```python
#import the required libraries
import bs4
import lxml
from bs4 import BeautifulSoup
import os
import json
```

```python
#navigate to the folder that contains the html files
os.chdir('html_files')
```

```python
tag_dict = {}
for file in os.listdir():
    
    if file.endswith("html"):
        print(file)
        f_dict = {}
        key_val = int(str(file)[0:-5])
        f = open(file,"r",encoding="utf-8")
        content = f.read()
        soup = BeautifulSoup(content,"lxml")

        #get title of document

        title = soup.find("div",{"class":"doc_title"})
        if(title):
            title = title.text
        else:
            title = ""


        #get author of document

        author = soup.find("div",{"class":"doc_author"})
        if(author):
            author = author.text
            author = author[8:] #to remove author keyword from the text
        else:
            author = ""

            
        f_dict["title"] = title
        f_dict["author"] = author

        
        
        
        para = {}
        ptext = ""
        #extract the paragraphs from the judgements
        for i in soup.find_all("p"):
            #f_dict[i.get('id')] = i.text
            ptext += i.text
            
        f_dict["content"] = ptext


        tag_dict[key_val] = f_dict
    
```

```python
#navigate back to the folder where you want to save the json file
os.chdir("..")
```

```python
with open('tag.json','w',encoding="utf-8") as outfile:
    json.dump(tag_dict,outfile,indent=4)
```
