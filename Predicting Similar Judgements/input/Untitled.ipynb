```python
def return_id(link):
    regex = re.compile(r'[0-9]+')
    res = regex.search(link)
    if res:
        return int(res.group())
```

```python
count = 0
for i in range(cases_21.shape[0]):
    main_j = return_id(cases_21.iloc[i,2])
    sim_j = return_id(cases_21.iloc[i,4])
    
    #get index from mapping
    
    if str(main_j) in mapping.keys() and str(sim_j) in mapping.keys():
        main_j_index = mapping[str(main_j)]
        sim_j_index = mapping[str(sim_j)]
        row = [main_j_index , 'similarTo' , sim_j_index]
        test_set.loc[count] = row
        count = count + 1
```
