```python
import pandas as pd
import numpy as np
import re
import json
```

```python
cases2020 = pd.read_csv("cases2020.csv")
```

```python
cases2021 = pd.read_csv("cases2021.csv")
```

```python
judgment_map = {}
```

```python
def get_id(link):
    regex = re.compile(r'[0-9]+')
    result = regex.search(link)
    if result:
        return int(result.group())
    else:
        return ''
```

```python
for i in range(cases2020.shape[0]):
    j = cases2020.iloc[i,0]
    j_link = cases2020.iloc[i,1]
    sj = cases2020.iloc[i,2]
    sj_link = cases2020.iloc[i,3]
    
    if len(j_link) > 2:
        j_id = get_id(j_link)
        if j_id not in judgment_map.keys():
            judgment_map[j_id] = j
            
    if len(sj_link) > 2:
        s_id = get_id(sj_link)
        if s_id not in judgment_map.keys():
            judgment_map[s_id] = sj
            
```

```python
judgment_map
```

```python
for i in range(cases2021.shape[0]):
    j = cases2021.iloc[i,0]
    j_link = cases2021.iloc[i,1]
    sj = cases2021.iloc[i,2]
    sj_link = cases2021.iloc[i,3]
    
    if not pd.isna(j_link):
        j_id = get_id(j_link)
        if j_id not in judgment_map.keys():
            judgment_map[j_id] = j
            
    if not pd.isna(sj_link):
        s_id = get_id(sj_link)
        if s_id not in judgment_map.keys():
            judgment_map[s_id] = sj
```

```python
judgment_map
```

```python
with open("JUDGMENT_ID_MAP.json", "w") as outfile:
    json.dump(judgment_map, outfile , indent=4)
```
