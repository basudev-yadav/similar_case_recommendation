```python
import numpy as np
import os
import pandas as pd
```

```python
stanza_files = pd.read_csv("stanza_triples_merged_csv.csv" , encoding = "utf-8")
```

```python
os_labels = pd.read_csv("OE_LABELS.csv" , encoding = "utf-8")
```

```python
stids = []
for i in range(stanza_files.shape[0]):
    stids.append(int(stanza_files.iloc[i,0]))

os_tids = []
for j in range(os_labels.shape[0]):
    os_tids.append(int(os_labels.iloc[j,0]))
    
```

```python
stids = np.unique(stids)
os_tids = np.unique(os_tids)
```

```python
len(stids), stids
```

```python
len(os_tids) , os_tids
```

```python
common_tids = []
for i in stids:
    if i in os_tids:
        common_tids.append(i)
len(common_tids)
```

```python
common_tids_dict = {}
for i in range(len(common_tids)):
    common_tids_dict[i] = int(common_tids[i])

```

```python
type(common_tids_dict)
```

```python
import json
outfile = open("common_tids_new.json" , "w")
json.dump(common_tids_dict , outfile)
```
