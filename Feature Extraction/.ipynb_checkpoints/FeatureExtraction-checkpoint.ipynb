```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv("Features.csv")
```

```python
file.columns
```

```python
features = []

for i in file.columns[:-1]:
    for j in file[i]:
        if not pd.isna(j):
            features.append(j)
```

```python
features
```

```python
features[13] = "Acquiescence"
```

```python
features
```

```python
for i in range(len(features)):
    x = features[i].lower()
    features[i] = x
```

```python
import re
```

```python
for i in range(len(features)):
    x = features[i]
    
    regex = re.compile(r'[a-zA-Z -]+')
    res = regex.search(x)
    
    if res:
        features[i] = res.group()
    else:
        features[i] = x
```
